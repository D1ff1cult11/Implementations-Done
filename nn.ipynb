```python
import numpy as np
dataset = np.genfromtxt("heart_failure_clinical_records_dataset.csv", delimiter=",", skip_header=1)

'''
anemia, high bp, diabetes, smoking: 0 = false
sex: 0 = woman
'''

x = dataset[:, :-1]
y = dataset[:, -1]
x = (x - x.min(axis = 0)) / (x.max(axis = 0) - x.min(axis = 0))

def relu(z):
  return np.maximum(0, z)
def sigmoid(z):
  g = 1 / (1 + np.exp(-z))
  return g

def diffsigmoid(z):
  s = sigmoid(z)
  return (s * (1 - s))
print(x.shape, y.shape)
print(x[0].shape, x[1].shape)
print(x.shape[1])
```

```python
def dense(ain, W, b, activation):
  units = W.shape[1]
  aout = np.zeros(units)
  zout = np.zeros(units)

  for j in range(0, units):
    wj = W[:,j]
    z = np.dot(wj, ain) + b[j]
    zout[j] = z
    aout[j] = activation(z)
  return aout, zout
```

```python
def sequential(xi, W1, b1, W2, b2):
    a1, z1 = dense(xi, W1, b1, sigmoid)
    a2, z2 = dense(a1, W2, b2, sigmoid)
    return a1, z1, a2, z2
```

```python

np.random.seed(1)

W1 = np.random.randn(12, 4) * 0.01
b1 = np.zeros(4)

W2 = np.random.randn(4, 1) * 0.01
b2 = np.zeros(1)



alpha = 0.05
epochs = 2000

for epoch in range(epochs):
  loss = 0

  dW1 = np.zeros_like(W1)
  dB1 = np.zeros_like(b1)
  dW2 = np.zeros_like(W2)
  dB2 = np.zeros_like(b2)

  for i in range(0, 299):
      xi = x[i]
      yi = y[i]

      a1, z1, a2, z2 = sequential(xi, W1, b1, W2, b2)
      yhat = a2[0]
      loss += -(yi*np.log(yhat) + (1-yi)*np.log(1-yhat))

      #backprop
      dz2 = yhat - yi
      dw2 = a1.reshape(-1, 1) * dz2
      db2 = np.array([dz2])

      dz1 = (W2[:, 0] * dz2) * diffsigmoid(z1)
      dw1 = xi.reshape(-1, 1) * dz1
      db1 = dz1

      dW2 += dw2
      dB2 += db2
      dW1 += dw1
      dB1 += db1

  W2 -= alpha * (dW2 / 299)
  b2 -= alpha * (dB2 / 299)
  W1 -= alpha * (dW1 / 299)
  b1 -= alpha * (dB1 / 299)

  if epoch % 200 == 0:
      print("epoch-", epoch, "cost-", loss/299)
```
