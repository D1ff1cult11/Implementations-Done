```python
X = df.drop(columns=60,axis=1)  #axis =1 column,axis=0 rows.
y = df[60]
```

```python
train_size =0.7
sample_rows = X.shape[0]
rows = np.random.permutation(sample_rows)
n = int(train_size*sample_rows)
test_idx = rows[n:]
train_idx = rows[:n]
```

```python
X_train =X.iloc[train_idx]#iloc gives permision to select rows and column, .iloc[selected_row,sel_col]  we can select rows location by slicing.
X_test =X.iloc[test_idx]
y_train =y.iloc[train_idx]
y_test =y.iloc[test_idx]
```

```python
def sigmoid(z):
  return 1/(1+np.exp(-z))

def loss(y,y_hat):
  m=y.shape[0] #no. of sample example.
  return (-1/m)* np.sum(y*np.log(y_hat+1e-9)+(1-y)*np.log(1-y_hat+1e-9))
```

```python
a,b=X.shape
W = np.zeros((b,1))
b=0

def train(X,y,lr=0.001,epochs=3000):

  for i in range(epochs):
    z=np.dot(X,W)+b
    y_hat = sigmoid(z)

    dw = (1/a) * np.dot(X.T,(y_hat-y))
    db = (1/a) * np.sum(y_hat - y)

    W -= lr * dw
    b -= lr * db

    if i%200 ==0:
      print(f"Epoch {i},Loss:{loss(y,y_hat):.4f}")
      # f: is formatting string(f-string)
      # which allows variable in string.

    return W,b
```

```python
def predict(X,W,b):
  probability = sigmoid(np.dot(X,W)+b)
  return (probability>=0.5).astype(int)#astype int convert True and False to 1 and 0(boolean).
```
