```python
import numpy as np
import pandas as pd
e=np.exp(-9)
def sigmoid(z):
  return 1 / (1 + np.exp(-z))
def softmax(z):
  exp_z=np.exp(z)
  return exp_z/np.sum(exp_z,axis=0)
def loss1(a):
  return np.mean(-y*np.log(a+e)-(1-y)*np.log(1-a+e))
def loss2(a3):
  return -np.sum(y*np.log(a3))
df = pd.read_csv("data.csv")
df.columns = ["grade1","grade2","label"]
x = df[["grade1","grade2"]]
y = df["label"].map(lambda x: float(x.rstrip(';')))
x = np.array(x)
y = np.array(y)
x[:,0] = (x[:,0] - np.mean(x[:,0]))/np.std(x[:,0])
x[:,1] = (x[:,1] - np.mean(x[:,1]))/np.std(x[:,1])
```

```python
W1=np.random.rand(3,2) * 0.8
b1=np.random.rand(3,1)*0.8
def dense(a_in,W,b):
  z=np.matmul(W,a_in)+b
  a_out=sigmoid(z)
  return a_out
W2=np.random.rand(4,3)*0.8
b2=np.random.rand(4,1)*0.8
W3=np.random.rand(1,4)*0.8
b3=np.random.rand(1,1)*0.8
epoches=1000
for epoch in range(epoches):

  z1 = np.matmul(W1, x.T) + b1
  a1 = sigmoid(z1)

  z2 = np.matmul(W2, a1) + b2
  a2 = sigmoid(z2)

  z3 = np.matmul(W3, a2) + b3
  a3 = sigmoid(z3)

  m=x.shape[1]
  y_reshaped = y.reshape(1, -1)


  dZ3 = a3 - y_reshaped
  dW3 = np.matmul(dZ3, a2.T) / m
  db3 = np.sum(dZ3, axis=1, keepdims=True) / m


  dZ2 = np.matmul(W3.T, dZ3) * (a2 * (1 - a2))
  dW2 = np.matmul(dZ2, a1.T) / m
  db2 = np.sum(dZ2, axis=1, keepdims=True) / m


  dZ1 = np.matmul(W2.T, dZ2) * (a1 * (1 - a1))
  dW1 = np.matmul(dZ1, x) / m
  db1 = np.sum(dZ1, axis=1, keepdims=True) / m

  W3 = W3 - 0.01 * dW3
  b3 = b3 - 0.01 * db3
  W2 = W2 - 0.01 * dW2
  b2 = b2 - 0.01 * db2
  W1 = W1 - 0.01 * dW1
  b1 = b1 - 0.01 * db1

def predict_model(X_input, W1, b1, W2, b2, W3, b3):
  z1 = np.matmul(W1, X_input.T) + b1
  a1 = sigmoid(z1)
  z2 = np.matmul(W2, a1) + b2
  a2 = sigmoid(z2)
  z3 = np.matmul(W3, a2) + b3
  a3 = sigmoid(z3)
  return (a3 >= 0.5).astype(int)

def accuracy_model(X_data, Y_true, W1, b1, W2, b2, W3, b3):
  y_pred = predict_model(X_data, W1, b1, W2, b2, W3, b3)
  y_true_reshaped = Y_true.reshape(1, -1)
  return np.mean(y_pred == y_true_reshaped)
loss_layer1=loss1(a1)
loss_layer2=loss1(a2)
loss_layer3=loss1(a3)

print(accuracy_model(x, y, W1, b1, W2, b2, W3, b3))
print(f"w={W1}")
print(f"b={b1}")
print(f"w={W2}")
print(f"b={b2}")
print(f"w={W3}")
print(f"b={b3}")
print(f"loss={loss_layer1}")
print(f"loss of layer 2:{loss_layer2}")
print(f"loss of layer 3:{loss_layer3}")
```
