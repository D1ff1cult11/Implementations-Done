```python
import numpy as np
import csv
fileid ="1vtQ442CfzInJrVkD3SwTOYUku8MEAmn6" #Here I have extracted this file from Google drive
!wget -O Heart_Disease_Prediction.csv "https://drive.google.com/uc?id=1vtQ442CfzInJrVkD3SwTOYUku8MEAmn6"
x_t=[]
y_t=[]
fi= open("Heart_Disease_Prediction.csv", "r")
reader= csv.reader(fi)
head= next(reader)
for r in reader:
  a=[]
  for k in r[:-1]:
    a.append(float(k))
  x_t.append(a)
  if r[-1]== "Presence":
    y_t.append(1)
  else:
    y_t.append(0)
x_train= np.array(x_t)
y_train=np.array(y_t)
#x_train= (x_train - np.mean(x_train, axis=0))/(np.max(x_train, axis=0)- np.min(x_train, axis=0)+ 1e-8)
```

```python
m= x_train.shape[1]
n= y_train.shape[0]
def sigmoid(z):
  s= 1/(1+np.exp(-z))
  return s
def relu(z):
  r= np.maximum(0,z)
  return r
def dsigmoid(z):
  s1= sigmoid(z)
  return (s1*(1-s1))
```

```python
def dense(a_in, W, b, g):
  units= W.shape[1]
  a_out=np.zeros(units)
  z_out= np.zeros(units)
  for j in range(units):
    w=W[:,j]
    z= np.dot(w, a_in) + b[j]
    a_out[j] = g(z)
    z_out[j]=z
  return (z_out, a_out)
```

```python
def sequential(x, W1, b1, W2, b2):
  z1, a1= dense(x, W1, b1, sigmoid)
  z2, a2= dense(a1, W2, b2, sigmoid)
  return (a1, a2, z1, z2)
```

```python
def predict(x, w1, w2, b1, b2):
  A1, A2, Z1, Z2 = sequential(x, w1, w2, b1, b2)
  if A2>=0.5:
    print("Disease present")
    return 1
  else:
    print("No disease")
    return 0
```

```python
import random
h=3
w1= np.random.randn(m,h)*0.001
b1= np.random.randn(h)*0.01
w2= np.random.randn(h,1)*0.01
b2= np.random.randn(1)*0.01
#w1= np.full((m, h), 0.001, dtype=float)
#b1= np.full(h, 0.001, dtype= float)
#w2= np.full((h,1), 0.001, dtype=float)
#b2= np.full(1, 0, dtype=float)
alpha=0.001
epochs=10000

for epoch in range(epochs):
  dw1= np.zeros_like(w1)
  dw2= np.zeros_like(w2)
  db1= np.zeros_like(b1)
  db2= np.zeros_like(b2)
  loss=0
  for i in range(n):

    xi= x_train[i]
    yi= y_train[i]
    #print(xi)
    a1, a2, z1, z2= sequential(xi, w1, b1, w2, b2)
    #print(w2.shape)
    loss+= -(yi*np.log(a2)+(1-yi)*np.log(1-a2))
    dz2= a2-yi
    dw2+=a1.reshape(-1,1)*dz2
    db2+=dz2
    dz1= w2.reshape(-1)*dz2*dsigmoid(z1)
    dw1+=xi.reshape(-1,1)*dz1
    db1+= dz1
  #print(w2.shape)
  w1-=alpha*dw1/n
  w2-=alpha*dw2/n
  b1-=alpha*db1/n
  b2-=alpha*db2/n
  loss/=n
  if epoch%500==0:
    print("Iteration no.:", epoch, "Loss=", loss)
print("W1 final:", w1, "W2 final:", w2, "b1 final:", b1, "b2 final:", b2)
```
