```python
import numpy as np
dataset = np.genfromtxt("heart_failure_clinical_records_dataset.csv", delimiter=",", skip_header=1)

'''
anemia, high bp, diabetes, smoking: 0 = false
sex: 0 = woman
'''
```

```python
x = dataset[:, :-1]
y = dataset[:, -1]
x = x / np.max(x)
```

```python
def sigmoid(z):
  z = np.clip(z, -500, 500)
  g = 1 / (1 + np.exp(-z))
  return g

def cost(x, y, w, b):

  m = 299
  cost = 0
  for i in range(0, m):
    z = np.dot(x[i], w) + b
    f_wb = sigmoid(z)
    f_wb = np.clip(f_wb, 1e-15, 1-1e-15)
    cost += (y[i] * np.log(f_wb)) + ((1 - y[i]) * np.log(1 - f_wb))

  tcost = -cost/m
  return tcost
```

```python
def grad(x, y, w, b):
  m = 299
  n = 12
  dj_dw = np.zeros(w.shape)
  dj_db = 0

  for i in range(0, m):
    f_wbi = sigmoid(np.dot(x[i], w) + b)
    errori = f_wbi - y[i]
    for j in range(0, n):
      dj_dw[j] = dj_dw[j] + errori * x[i, j]
    dj_db = dj_db + errori
  dj_dw = dj_dw / m
  dj_db = dj_db / m

  return dj_db, dj_dw
```

```python
def grad_desc(x, y, w, b, epochs, alpha):
  m = 299
  n = 12
  all_j = []
  all_w = []

  for i in range(epochs):
    dj_db, dj_dw = grad(x, y, w, b)

    w = w - alpha * dj_dw
    b = b - alpha * dj_db

    tcost = cost(x, y, w, b)

    all_j.append(tcost)
    all_w.append(w)
    if (i % 100 == 0):
      print("epoch:", i, "Cost:", tcost)

  return w, b, all_j, all_w
```

```python
np.random.seed(1)
w_initial = 0.01 * (np.random.rand(12).reshape(-1, 1) - 0.5)
b_initial = 0.001
epochs = 1000
alpha = 0.00001

w, b, all_j, all_w = grad_desc(x, y, w_initial, b_initial, epochs, alpha)
print("w- ", w, "b- ", b)
```
