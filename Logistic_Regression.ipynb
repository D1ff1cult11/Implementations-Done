```python
import csv
x_t=[]
y_t=[]
fi= open("Heart_Disease_Prediction.csv", "r")
reader= csv.reader(fi)
head= next(reader)
for r in reader:
  a=[]
  for k in r[:-1]:
    a.append(float(k))
  x_t.append(a)
  if r[-1]== "Presence":
    y_t.append(1)
  else:
    y_t.append(0)
x_train= np.array(x_t)
y_train=np.array(y_t)
```

```python
m= x_train[0].shape[0]
w= np.zeros(m)
n= y_train.shape[0]
b=0
z=0
def computez(w,b,xi):
    z= np.dot(w,xi) + b
    return z
```

```python
def computef(z):
    e= np.exp(-z)
    f= 1/(1+e)
    return f
```

```python
import math
def computecost(wu,bu):
  cost= 0
  for i in range(n):
    xi= x_train[i]
    yi= y_train[i]
    z= np.dot(wu,xi)+b
    fu= computef(z)
    #if fu>1 or fu<0:
      #print(fu)
      #print(xi)
    if fu==0:
      fu+=0.0000000000001
    if fu==1:
      fu-=0.0000000000001
    #fu*=0.1
    #print(fu)
    cost+= -yi*np.log(fu) - (1-yi)*np.log(1-fu)
  cost/=n

  #print(wu, bu, cost)
  return cost
```

```python
def compute_gradient(w,b):
  dj_dw= np.zeros(m)
  dj_db=0
  for i in range(n):
    xi= x_train[i]
    yi= y_train[i]
    z= np.dot(w,xi)+b
    f= computef(z)
    gr= f-yi
    dj_dw+= gr*xi
    dj_db= gr
  dj_dw/=n
  dj_db/=n
  return dj_dw, dj_db
```
